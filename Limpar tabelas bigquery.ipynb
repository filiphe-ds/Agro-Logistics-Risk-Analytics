```python
from google.cloud import bigquery

client = bigquery.Client(project="agrologisticsdata")
dataset_id = "logisticsdata"
tabelas = ["fato_clima", "fato_lineup", "fato_contingencias_nlp"]

for tabela in tabelas:
    table_ref = f"{client.project}.{dataset_id}.{tabela}"
    try:
        client.delete_table(table_ref, not_found_ok=True)
        print(f"🗑️ Tabela {tabela} removida com sucesso.")
    except Exception as e:
        print(f"❌ Erro ao deletar {tabela}: {e}")

# Depois disso, basta rodar seus scripts originais de criação de tabela (CREATE TABLE).
```
